```python
df.loc[df.ocorrencia_uf == '***', ['ocorrencia_uf']] = pd.NA #Realiza a alteração conforme parametros informados de coluna e valor procurado, para retornar <NA>
```

```python
df.replace(['***', 'NULL'], pd.NA, inplace=True) #Realiza alteração em todos os campos que encontrarem determinado valor, independentemente de onte estejam no datafram o comundo inplace=True indifica que a alteração é realmente no dataframe se não colocar, fica somente para visualuzação
```

```python
df.isna().sum() #É uma maneira de tratar os dados não informando, esta linha, fas uma contagem dos dados NA nas colunas
```

```python
df. isnull().sum() #É uma maneira de tratar os dados não informando, esta linha, fas uma contagem dos dados NA nas colunas
```

```python
df.fillna(0) #Altera os dados não informando para o parametro indicado, no caso "0", repare que está só visualização, mas se quiser trocar no dataframe, preciso inserir a função inplace
```

```python
df.fillna(value={'ocorrencia_uf':'BR'}) #Altera somente os valores NA da coluna informada para o indicado, no caso alterou os valores da coluna ocorrencias UF que eram NA para BR
```
